```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Titanic-Dataset.csv')


print("=== Summary Statistics ===")
print(df.describe(include='all'))


numeric_cols = df.select_dtypes(include=np.number).columns


df[numeric_cols].hist(bins=15, figsize=(15, 10), layout=(len(numeric_cols) // 3 + 1, 3))
plt.tight_layout()
plt.suptitle('Histograms of Numeric Features', fontsize=16)
plt.subplots_adjust(top=0.95)
plt.show()


for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df, x=col)
    plt.title(f'Boxplot of {col}')
    plt.show()

sns.pairplot(df[numeric_cols])
plt.suptitle('Pairplot of Numeric Features', fontsize=16, y=1.02)
plt.show()

plt.figure(figsize=(12, 8))
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()


for col in numeric_cols:
    outliers = df[(df[col] < (df[col].quantile(0.25) - 1.5 * (df[col].quantile(0.75) - df[col].quantile(0.25)))) |
                  (df[col] > (df[col].quantile(0.75) + 1.5 * (df[col].quantile(0.75) - df[col].quantile(0.25))))]
    if not outliers.empty:
        print(f"\nPotential outliers detected in '{col}':")
        print(outliers[[col]])


print("\n=== Basic Inferences from Correlations ===")
high_corr_pairs = corr.abs().unstack().sort_values(ascending=False)
high_corr_pairs = high_corr_pairs[high_corr_pairs < 1].drop_duplicates()
print(high_corr_pairs[high_corr_pairs > 0.6])
```
